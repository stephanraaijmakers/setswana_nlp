```python
import numpy as np
import re
from transformers import AutoTokenizer, AutoModel
from bertviz import head_view,model_view

tokenizer = AutoTokenizer.from_pretrained("/Users/stephan/VIRTUALENVS/transformers/code/tswana_models/output")
model = AutoModel.from_pretrained("/Users/stephan/VIRTUALENVS/transformers/code/tswana_models/output", output_attentions=True)

with open("/Users/stephan/VIRTUALENVS/transformers/code/all.sentences.txt","r") as fp:
    lines=fp.readlines()
    
n=1
max_prob=0.0

winner=""
for line in lines[33000:33100]:
    line =re.sub("[\.\,\:\;]","",line.rstrip())
    if n%100==0: 
        print("Line:%d"%(n)) 
    words=line.split(" ")
    if len(words)<5:
        continue
    inputs = tokenizer.encode(line, return_tensors='pt')
    outputs = model(inputs)
    attention = outputs[-1]  # Output includes attention weights when output_attentions=True
    for i in range(attention[0].shape[0]):
        w=attention[0][i].detach().numpy()
        for l in w: #w[-1:]: # last layer
            for weights in l:
                for v in weights:
                    if v>max_prob:
                        winner=line
                        print("Current winner:%s (%f)"%(winner,v))
                        max_prob=v
    n+=1
    
# Dilo tse di maswe ka tota

# Bontsi bo ne bo tlile go keteka letsatsi le le bobedi jo
# Many would come to celebrate the day
# Bontsi=many 
# bo ne bo = were => Most were (flexie, agreement
# tlile:come
# go keteka: to celebrate
# 

if winner!="":
    inputs = tokenizer.encode(winner, return_tensors='pt')
    outputs = model(inputs)
    attention = outputs[-1]                        
    tokens = tokenizer.convert_ids_to_tokens(inputs[0])
    head_view(attention, tokens)
    
    
    
    
```
